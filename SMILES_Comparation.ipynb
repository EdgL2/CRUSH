```python
# Librerías
import pandas as pd
from google.colab import files
from rdkit import Chem
from itertools import combinations
```

```python
# ============================================================
# FUNCIÓN PARA CANONICALIZAR SMILES
# ============================================================

def canonicalize_smiles(smiles):
    """
    Convierte un SMILES a formato canónico usando RDKit.
    Si el SMILES es inválido, devuelve None.
    """
    try:
        mol = Chem.MolFromSmiles(str(smiles))
        if mol is not None:
            return Chem.MolToSmiles(mol, canonical=True)
        else:
            return None
    except:
        return None
```

```python
# ============================================================
# DICCIONARIOS PARA ALMACENAR DATAFRAMES
# ============================================================

dfs = {}
sets_smiles = {}
```

```python
# ============================================================
# CARGAR 5 ARCHIVOS UNO A LA VEZ
# ============================================================

for i in range(1, 6):

    print("\n" + "="*60)
    print(f"CARGAR ARCHIVO {i} DE 5")
    print("="*60)

    uploaded = files.upload()

    # Obtener nombre del archivo
    filename = list(uploaded.keys())[0]

    print(f"\nProcesando: {filename}")

    # --------------------------------------------------------
    # LEER ARCHIVO (VERSIÓN ROBUSTA)
    # --------------------------------------------------------

    if filename.endswith(".csv"):

        try:
            # Intento estándar
            df = pd.read_csv(filename)

        except:

            try:
                # Intento con ;
                df = pd.read_csv(filename, sep=";")

            except:

                try:
                    # Intento automático
                    df = pd.read_csv(
                        filename,
                        sep=None,
                        engine="python"
                    )

                except Exception as e:

                    print(f"No se pudo leer el archivo: {e}")
                    continue

    elif filename.endswith(".xlsx"):

        df = pd.read_excel(filename)

    elif filename.endswith(".tsv"):

        df = pd.read_csv(filename, sep="\t")

    else:

        print("Formato no soportado")
        continue

    # --------------------------------------------------------
    # VERIFICAR COLUMNA SMILES
    # --------------------------------------------------------

    if "SMILES" not in df.columns:

        print("El archivo no contiene una columna llamada 'SMILES'")
        continue

    # --------------------------------------------------------
    # LIMPIAR Y CANONICALIZAR
    # --------------------------------------------------------

    df = df.dropna(subset=["SMILES"]).copy()

    df["SMILES_canonical"] = df["SMILES"].apply(canonicalize_smiles)

    # Eliminar inválidos
    df = df.dropna(subset=["SMILES_canonical"])

    # --------------------------------------------------------
    # GUARDAR
    # --------------------------------------------------------

    db_name = filename.split(".")[0]

    dfs[db_name] = df

    sets_smiles[db_name] = set(df["SMILES_canonical"])

    # --------------------------------------------------------
    # RESUMEN
    # --------------------------------------------------------

    print(f"Moléculas válidas: {len(df)}")

    print(f"SMILES únicos: {len(sets_smiles[db_name])}")
```

```python
# ============================================================
# COMPARACIÓN ENTRE BASES DE DATOS
# ============================================================

print("\n" + "="*70)
print("COMPARACIÓN DE COMPUESTOS REPETIDOS")
print("="*70)

results_summary = []

for db1, db2 in combinations(sets_smiles.keys(), 2):

    repeated = sets_smiles[db1].intersection(sets_smiles[db2])

    print(f"\n{db1} vs {db2}")

    print(f"Compuestos repetidos: {len(repeated)}")

    results_summary.append({
        "Database_1": db1,
        "Database_2": db2,
        "Repeated_compounds": len(repeated)
    })
```

```python
# ============================================================
# MATRIZ DE INTERSECCIONES
# ============================================================

db_names = list(sets_smiles.keys())

matrix = pd.DataFrame(index=db_names, columns=db_names)

for db1 in db_names:

    for db2 in db_names:

        repeated = sets_smiles[db1].intersection(sets_smiles[db2])

        matrix.loc[db1, db2] = len(repeated)

print("\n" + "="*70)
print("MATRIZ DE COMPUESTOS COMPARTIDOS")
print("="*70)

display(matrix)
```

```python
# ============================================================
# IDENTIFICAR TODOS LOS COMPUESTOS REPETIDOS
# ============================================================

all_smiles = {}

for db_name, smiles_set in sets_smiles.items():

    for smi in smiles_set:

        if smi not in all_smiles:

            all_smiles[smi] = []

        all_smiles[smi].append(db_name)
```

```python
# ------------------------------------------------------------
# CREAR DATAFRAME
# ------------------------------------------------------------

repeated_data = []

for smi, dbs in all_smiles.items():

    if len(dbs) > 1:

        repeated_data.append({
            "SMILES": smi,
            "Databases": ", ".join(dbs),
            "Count_databases": len(dbs)
        })

df_repeated = pd.DataFrame(repeated_data)

# Ordenar
df_repeated = df_repeated.sort_values(
    by="Count_databases",
    ascending=False
)

print("\n" + "="*70)
print("COMPUESTOS PRESENTES EN MÁS DE UNA BASE")
print("="*70)

display(df_repeated)
```

```python
# ============================================================
# GUARDAR RESULTADOS
# ============================================================

matrix.to_csv("Shared_compounds_matrix.csv")

df_repeated.to_csv(
    "Repeated_compounds_detailed.csv",
    index=False
)

print("\nArchivos generados:")
print("1. Shared_compounds_matrix.csv")
print("2. Repeated_compounds_detailed.csv")

# Descargar
files.download("Shared_compounds_matrix.csv")
files.download("Repeated_compounds_detailed.csv")
```
